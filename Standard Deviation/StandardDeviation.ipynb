# Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased',axis=1),df['Purchased'],test_size=0.3,random_state=0)

X_train.shape , X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(X_train) #Scaler is finding standard deviation and mean from train data but will transform both

X_train_scaled = scaler.transform(X_train) #Doing Standard Deviation
X_test_scaled = scaler.transform(X_test)
X_train_scaled[0:10]
```

```python
scaler.mean_
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled , columns=X_train.columns) # By default we get an array but we want a dataframe so we convert it to one
X_test_scaled = pd.DataFrame(X_test_scaled , columns=X_test.columns)
X_train_scaled
```
