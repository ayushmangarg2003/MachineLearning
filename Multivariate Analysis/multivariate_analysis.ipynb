```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/content/SharkTankIndiaDataset.csv')
df.head(2)
```

# Scatter Plot (Numerical - Numerical)

```python
sns.scatterplot(x=df['ask_equity'], y=df['deal_equity'],hue=df['peyush_deal'],style=df['aman_deal'],size=df['ask_valuation'])
```

# Bar Plot (Numerical - Categorical)

```python
sns.barplot(x=df['total_sharks_invested'],y=df['deal_amount'],hue=df['aman_deal'])
```

# Box Plot (Numerical - Categorical)

```python
sns.boxplot(x=df['total_sharks_invested'],y=df['deal_amount'])
```

# Heat Map (Categorical - Categorical)

```python
 sns.heatmap(pd.crosstab(df['total_sharks_invested'],df['aman_deal']))
```

# Cluster map (Categorical - Categorical)

```python
 sns.clustermap(pd.crosstab(df['total_sharks_invested'],df['aman_deal']))
```

# Pair Plot

```python
sns.pairplot(df)
```
